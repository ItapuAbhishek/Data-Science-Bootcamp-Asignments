The dataset

Download the dataset here (Source: UC Irvine Machine Learning Repository).

This dataset is composed of a range of biomedical voice measurements from 31 people, 23 with Parkinson's disease (PD). Each column in the table is a particular voice measure, and each row corresponds to one of 195 voice recordings from these individuals (the name column). The main aim of the data is to discriminate healthy people from those with PD, according to the status column, which is set to 0 for healthy and 1 for PD.

For this exercise, you need the following:

pandas
train_test_split
LogisticRegression
PyPlot from Matplotlib

```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

2. Load and prepare the dataset

Load the dataset into a DataFrame named df_data.
Drop the name column. This column is not a feature required by the model.
Create a DataFrame named X containing all the columns except the status.
Create a DataFrame named y containing the status.

```
df=pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/parkinsons.data')

df.info()
df=df.drop(['name'],axis=1)
y=df['status']
X=df.drop(['status'],axis=1)
```

3. Split the dataset into a training set and a test set

```
x_train,x_test,y_train,y_test=train_test_split(X,y)
```

4. Create the model

Instantiate a logistic regression classifier with an lbfgs solver and penalty set to none.
Fit the classifier to the data.
If the algorithm fails to converge, what can you change to fix that?

```
lr= LogisticRegression(solver='lbfgs',penalty='none',max_iter=1000,random_state=123)
lr.fit(x_train,y_train)
```

5. Calculate accuracy

Calculate and print the accuracy of the model on the test data.

```
print('accuracy score {}'.format(lr.score(x_test,y_test)))
```

6. L2 regularization

Train another model with L2 regularization and compare the accuracy of the new model to the previous one.

```
lr1= LogisticRegression(solver='lbfgs',penalty='l2',max_iter=1000,random_state=12345)
lr1.fit(x_train,y_train)
print('Accuracy score {}'.format(lr1.score(x_test,y_test)))
```
